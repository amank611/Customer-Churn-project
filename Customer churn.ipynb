```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Calculate counts
counts = ab.groupby(['SeniorCitizen', 'Churn']).size().unstack()

# Convert to percentages
counts_perc = counts.div(counts.sum(axis=1), axis=0) * 100

# Plot stacked bar chart
ax = counts_perc.plot(kind='bar', stacked=True, figsize=(6,4))

# Add percentage labels
for c in ax.containers:
    ax.bar_label(c, fmt='%.1f%%', label_type='center', fontsize=10)

plt.title("Churn by SeniorCitizen (Stacked %)")
plt.ylabel("Percentage")
plt.xlabel("Senior Citizen")
plt.legend(title="Churn",bbox_to_anchor=(.9,.9))

plt.xticks(rotation=0)
plt.show()
```

```python

# Define column names
columns = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Define number of rows and columns for subplots
rows = 3
cols = 3

# Create subplots
fig, axes = plt.subplots(rows, cols, figsize=(15, 12))  # Adjust figure size

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Loop through columns and plot countplots
for i, col in enumerate(columns):
    ac=sns.countplot(x=col, data=ab, ax=axes[i],hue='Churn')  # Use seaborn countplot
    ac.bar_label(ac.containers[0])
    axes[i].set_title(f"Distribution of {col}")  # Set title
    axes[i].set_xlabel("")  # Remove x-label for cleaner look
    axes[i].set_ylabel("Count")

# Adjust layout
plt.tight_layout()
plt.show()
```

#Key insights include a higher churn rate among customers with no security, backup, or tech support services.Fiber optic internet users also show a notable churn rate compared to DSL users. Additionally, customers without phone services have a lower count but a relatively higher churn percentage.
